```python
import sympy as sp

# Z-transform of unit step u[n]
def z_transform_unit_step():
    z = sp.symbols('z')
    U_z = 1 / (1 - (1/z))  # Z-transform
    return U_z

# Main
if __name__ == "__main__":
    U = z_transform_unit_step()
    print("Z-transform of unit step u[n]:", U)
    print("ROC: |z| > 1 → Unit step is UNSTABLE")
```

```python
import sympy as sp

# System H(z) = 0.5((z - 0.7)(z - 0.9)) / ((z - 0.6)(z - 0.4))
z = sp.symbols('z')
H_z = 0.5 * (z - 0.7) * (z - 0.9) / ((z - 0.6) * (z - 0.4))
print("H(z) =", H_z)

# Find poles
denominator = (z - 0.6) * (z - 0.4)
poles = sp.solve(denominator, z)
print("Poles of H(z):", poles)

# Check stability
stable = all(abs(p) < 1 for p in poles)

if stable:
    print("The system is STABLE (All poles lie inside the unit circle)")
else:
    print("The system is UNSTABLE (Pole lies outside the unit circle)")
```
