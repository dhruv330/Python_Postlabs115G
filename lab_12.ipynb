```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)

sine1 = np.sin(2 * np.pi * 5 * t)
sine2 = np.sin(2 * np.pi * 10 * t)

added_signal = sine1 + sine2

plt.figure(figsize=(10, 4))
plt.plot(t, added_signal)
plt.title("Added Signal (5 Hz + 10 Hz Sine Waves)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 500
t = np.linspace(0, 2, 2*fs, endpoint=False)

sine_wave = np.sin(2 * np.pi * 5 * t)
cosine_wave = np.cos(2 * np.pi * 10 * t)

multiplied_signal = sine_wave * cosine_wave

plt.figure(figsize=(10, 4))
plt.plot(t, multiplied_signal)
plt.title("Product of 5 Hz Sine and 10 Hz Cosine")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)

sine_wave = np.sin(2 * np.pi * 5 * t)
shifted_wave = np.sin(2 * np.pi * 5 * (t - 0.1))

plt.figure(figsize=(10, 4))
plt.plot(t, sine_wave, label="Original 5 Hz Sine")
plt.plot(t, shifted_wave, label="Shifted by 0.1s", linestyle="--")
plt.title("Original vs Time-Shifted Sine Wave (5 Hz)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)

original_wave = np.sin(2 * np.pi * 10 * t)
scaled_wave = 3 * original_wave

plt.figure(figsize=(10, 4))
plt.plot(t, original_wave, label="Original 10 Hz Sine")
plt.plot(t, scaled_wave, label="Scaled by 3", linestyle="--")
plt.title("Original vs Scaled Sine Wave (10 Hz)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)

original_wave = np.sin(2 * np.pi * 5 * t)
reversed_wave = original_wave[::-1]

plt.figure(figsize=(10, 4))
plt.plot(t, original_wave, label="Original 5 Hz Sine")
plt.plot(t, reversed_wave, label="Reversed in Time", linestyle="--")
plt.title("Original vs Time-Reversed Sine Wave (5 Hz)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```
